```python
import scipy as sp
import matplotlib.pyplot as plt
%pylab inline
```

```python
X1 = []
X2 = []

with open("../result/1T.txt", "r") as file:
    for line in file:
        x1, x2 = [float(part.strip()) for part in line.strip().split(",")]
        X1.append(x1)
        X2.append(x2)

t = sp.linspace(0, 10, num=len(X1), endpoint=True)

        
fig = plt.figure(figsize=(15,6))
plt.plot(t, X1, label="x1")
plt.plot(t, X2, label="x2")
plt.title("Trapezni postupak X=(1,1)")
plt.xlabel("t - iteracija")
plt.legend(loc='best')
plt.grid()
plt.show()


###############################################

X1 = []
X2 = []

with open("../result/1RK.txt", "r") as file:
    for line in file:
        x1, x2 = [float(part.strip()) for part in line.strip().split(",")]
        X1.append(x1)
        X2.append(x2)
        
fig = plt.figure(figsize=(15,6))
plt.plot(t, X1, label="x1")
plt.plot(t, X2, label="x2")
plt.title("RungeKutta postupak X=(1,1)")
plt.xlabel("t - iteracija")
plt.legend(loc='best')
plt.grid()
plt.show()
```
